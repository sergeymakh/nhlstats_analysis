- В течение игры - шлём сообщение, если кто-то из российских игроков набирает очки
- После заверешния матча высылаем результата матча и полную статистику наших хоккеистов
2. 1 раз в сутки вечером высылать анонс предстоящих матчей на ближайшие 24 часа с участием российских хоккеистов
3. 1 раз в сутки высылать инфу по статистике:
- топ 10 по очкам (выделяем жирным наших)
- топ 10 по шайбам (аналогично)
- топ 10 по ассистам (аналогично)
- Статистику всех наших игроков (очки-шайбы-ассисты-плюс-минус-позиция) из топ 100 бомбардиров
